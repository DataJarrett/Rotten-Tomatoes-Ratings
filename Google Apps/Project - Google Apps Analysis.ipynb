# Data cleaning

```python
#BEFORE
apps = pd.read_csv('googleplaystore.csv')
apps.head()
```

```python
#CLEANED
def Categoryclean():
    apps.Category.fillna(value="ART_AND_DESIGN", inplace=True)

def Typeclean():
    apps.Type.fillna(value="Free", inplace=True)
    
def Genreclean():
    apps.Genres.fillna(value="Art & Design", inplace=True)

def InstallInteger():
    apps.Installs = apps.Installs.str.replace('+', '')
    apps.Installs = apps.Installs.str.replace(',', '').astype(int) 
    apps.Reviews = apps.Reviews.astype(int)

def DeviceClean():
    apps['Current Ver'] = apps['Current Ver'].replace("Varies with device", np.nan)
    apps['Android Ver'] = apps['Android Ver'].replace("Varies with device", np.nan)
    apps.Size = apps.Size.replace("Varies with device", np.nan)

    
#def SizeFloat():
 #  apps.Size = apps.Size.str.replace("M", "")
 #   KValues = apps[apps.Size.str.endswith("k")]
  #  KValues.Size.str.replace("k", "").astype(float).div(1000)
   # apps.Size = apps.Size.replace("Varies with device", np.nan)
  
    
def PriceInt():
    apps.Price = apps.Price.str.replace('$', '').astype(float)
    
def SizeFloat():    
    M_rows = apps.Size.str.contains('M')
    k_rows = apps.Size.str.contains('k')

    apps.loc[M_rows, 'Size'] = apps.loc[M_rows, 'Size'].str.replace('M', "").astype('float')
    apps.loc[k_rows, 'Size'] = apps.loc[k_rows, 'Size'].str.replace('k', "").astype('float').div(1000).round(2)

apps = pd.read_csv('googleplaystore.csv')
Categoryclean() #fix missing
Typeclean() #fix missing
Genreclean() #fix missing
InstallInteger() #Integer Installs
PriceInt() #Integer Prices
SizeFloat() #formats size into float
DeviceClean()#remove Varies with device issues

#apps = apps.dropna()
apps = apps.drop_duplicates("App")
apps = apps.reset_index()
apps = apps.drop(columns=['index', 'Last Updated', 'Current Ver', 'Android Ver'])

#version formats could turn into a number so you can have some sort of comparison? otherwise useless
```

```python
#The question: Does more installs mean higher rating? I.e. Does rating matter?
```

```python
gb = apps.groupby('Category')
Ratings = gb[['Rating', 'Installs']].mean().round(2)
Ratings
```

```python
#hm = Ratings.plot(kind='scatter', x='Installs', y='Rating', figsize=(8,5), marker='*')
hm = Ratings.plot(kind='scatter', x='Installs', y='Rating', logx=True, figsize=(8,5), marker='*')
hm.axhline(y=4, color='r') 
```

```python
# So no, your rating does not drastically affect the number of installs you achieve. 
#You can have very few downloads but very high ratings - however, it does show that higher installs does lead to higher ratings
#may saturate which then brings negative reviews? >i.e. they tend towards 4.15
```

```python
Ratings.corr()
```
